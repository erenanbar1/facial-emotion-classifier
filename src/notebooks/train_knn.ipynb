```python
import sys
sys.path.append("../scripts")  # path to folder
from create_train_test_dicts import main

train_dict, numeric_train, test_dict, numeric_test = main()
```

```python
from knn import find_best_k, train_knn, test_knn 

best_k, best_score, scores = find_best_k(numeric_train,
                                         k_values=range(1, 26, 2),  # 1..25 odd
                                         cv=5)

print("Best k:", best_k)
print("Best Cross-Validation accuracy:", best_score)


#train

knn_model, le = train_knn(numeric_train, n_neighbors=5)

```

```python
#test
cm, class_names = test_knn(knn_model, le, numeric_test)
print("Class order:", class_names)
print("Confusion matrix:\n", cm)
```

```python
from metrics_from_cm import metrics_from_confusion_matrix
results = metrics_from_confusion_matrix(cm)

print("Accuracy:", results["accuracy"])
print("Macro Precision:", results["macro_precision"])
print("Macro Recall:", results["macro_recall"])
print("Macro F1:", results["macro_f1"])

print("\nPer-class metrics:")
for cls, p, r, f in zip(class_names, 
                        results["precision_per_class"],
                        results["recall_per_class"],
                        results["f1_per_class"]):
    print(f"{cls:10s}  P={p:.3f}  R={r:.3f}  F1={f:.3f}")
```
