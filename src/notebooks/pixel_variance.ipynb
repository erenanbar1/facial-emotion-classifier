```python
import sys
sys.path.append("../")
from preprocessing.create_train_test_dicts import build_splits
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Load the train split dictionary
# -----------------------------
ds = build_splits()
train_split = ds['train_numeric']
# -----------------------------
# Convert dict → huge (N, 48, 48) array
# -----------------------------
all_images = []
for lbl, imgs in train_split.items():
    all_images.extend(imgs)     # imgs = list of 48×48 arrays

all_images = np.array(all_images)  # shape: (N, 48, 48)
print("Dataset shape:", all_images.shape)

# -----------------------------
# Compute pixel-wise variance
# -----------------------------
pixel_variance = all_images.var(axis=0)  # shape: (48, 48)

# -----------------------------
# Visualize variance heatmap
# -----------------------------
plt.figure(figsize=(5,5))
plt.imshow(pixel_variance, cmap='hot')
plt.colorbar(label="Pixel-wise variance")
plt.title("FER2013 Pixel Variance Across Training Images")
plt.show()
```
